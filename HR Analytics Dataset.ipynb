#### Importing Libraries

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
```

```python
train = pd.read_csv('train_LZdllcl.csv', index_col = 'employee_id')
test = pd.read_csv('test_2umaH9m.csv', index_col = 'employee_id')
```

#### Data Processing

```python
train['previous_year_rating'].fillna(train['previous_year_rating'].mean(), inplace = True)
```

#### Converting Categorical Values into Numeric Values

```python
train['recruitment_channel'] = train['recruitment_channel'].replace({'other':0, 'sourcing':1, 'referred':2})
```

```python
test['recruitment_channel'] = test['recruitment_channel'].replace({'other':0, 'sourcing':1, 'referred':2})
```

```python
train.head()
```

```python
train['gender'].value_counts()
```

```python
train['gender'] = train['gender'].replace({'f':0, 'm':1})
test['gender'] = test['gender'].replace({'f':0, 'm':1})
```

```python
train.tail()
```

```python
train.drop('region', axis = 1, inplace = True)
test.drop('region', axis = 1, inplace = True)
```

```python
train.head()
```

```python
train.department.value_counts()
```

```python
train['department'] = train['department'].replace({'Sales & Marketing':0, 'Operations':1, 'Technology':2, 'Procurement':3, 'Analytics':4, 'Finance':5, 'HR':6, 'Legal':7, 'R&D':8})
```

```python
test['department'] = test['department'].replace({'Sales & Marketing':0, 'Operations':1, 'Technology':2, 'Procurement':3, 'Analytics':4, 'Finance':5, 'HR':6, 'Legal':7, 'R&D':8})
```

```python
train.head()
```

```python
test.head()
```

```python
train.education.value_counts()
```

```python
train.drop('education', axis = 1, inplace = True)
test.drop('education', axis = 1, inplace = True)
```

```python
train.head()
```

```python
test.tail()
```

```python
features = ['department', 'gender', 'recruitment_channel', 'no_of_trainings', 'age', 'previous_year_rating', 'length_of_service', 'KPIs_met >80%', 'awards_won?', 'avg_training_score']

X_train = train[features]
Y_train = train['is_promoted']
X_test = test
```

```python
X_train.shape, Y_train.shape, X_test.shape
```

#### Creating Models for Predictions

```python
#Using Random Forest Algorithm
model = RandomForestRegressor(n_estimators = 100)
model.fit(X_train, Y_train)
predicted = model.predict(X_test)
predicted = predicted.round(0).astype(int)
```

```python
#Using Decision Tree Algorithm
clf = DecisionTreeRegressor()
clf.fit(X_train, Y_train)
pred = clf.predict(X_test)
pred = pred.round(0).astype(int)
```

```python
print('Training Score using Random Forest: ', model.score(X_train, Y_train))
print('Testing Score using Random Forest: ', model.score(X_test, predicted))
```

```python
print('Training Score using Decision Tree: ', clf.score(X_train, Y_train))
print('Testing Score using Decision Tree: ', clf.score(X_test, pred))
```

```python
output = pd.DataFrame({'employee_id': test.index, 'is_promoted': pred})
output.to_csv('submission.csv', index = False)
```
